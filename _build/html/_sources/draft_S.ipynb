```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr
import plotly.express as px
```

```python
# Load the dfset
file_path = 'filtered_global_education_world_happiness_average.csv'
df = pd.read_csv(file_path)
```

```python
# Calculate the Average Completion Rate
df['Average_Completion_Rate'] = df[['Average_Completion_Rate_Male', 'Average_Completion_Rate_Female']].mean(axis=1)

# Filter out entries with 0 values in the Average Completion Rate
filtered_data = df[df['Average_Completion_Rate'] != 0]

# Calculate the Average Proficiency (Math and Reading)
df['Average_Proficiency'] = df[['Lower_Secondary_End_Proficiency_Reading', 'Lower_Secondary_End_Proficiency_Math']].mean(axis=1)

# Filter out entries with 0 values in the Average Proficiency
filtered_proficiency_data = df[df['Average_Proficiency'] != 0]

# Calculate Pearson correlation coefficients and p-values
r_completion, p_value_completion = pearsonr(filtered_data['Average_Completion_Rate'], filtered_data['Ladder score'])
r_proficiency, p_value_proficiency = pearsonr(filtered_proficiency_data['Average_Proficiency'], filtered_proficiency_data['Ladder score'])

# Create a combined scatter plot with regression lines for both variables with improved aesthetics
plt.figure(figsize=(12, 8))

# Plot Average Completion Rate and Happiness
sns.regplot(data=filtered_data, x='Average_Completion_Rate', y='Ladder score', 
            scatter_kws={'alpha': 0.5, 's': 60}, line_kws={'color': 'steelblue'}, 
            label='Average Completion Rate', color='skyblue')

# Plot Average Proficiency and Happiness
sns.regplot(data=filtered_proficiency_data, x='Average_Proficiency', y='Ladder score', 
            scatter_kws={'alpha': 0.5, 's': 60}, line_kws={'color': 'darkseagreen'}, 
            label='Average Proficiency (Math and Reading)', color='darkseagreen')

# Add titles and labels with improved aesthetics
plt.title('Relationship Between Education Metrics and Happiness', fontsize=16)
plt.xlabel('Education Metrics', fontsize=14)
plt.ylabel('Happiness (Ladder Score)', fontsize=14)
plt.legend(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)

# Set the style of the plot to improve aesthetics
sns.set_style("whitegrid")

# Display correlation coefficients and p-values in the plot, aligned left and below the x-axis title
plt.gca().text(-0.05, -0.2, 
            f'Pearson correlation: r = {r_completion:.2f}, p-value = {p_value_completion:.2e}\n'
            f'Pearson correlation: r = {r_proficiency:.2f}, p-value = {p_value_proficiency:.2e}',
            fontsize=12, bbox=dict(facecolor='white', alpha=0.5), transform=plt.gca().transAxes, ha='left')

plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from IPython.display import display, clear_output
import ipywidgets as widgets

# Load the dataset
file_path = 'filtered_global_education_world_happiness_average.csv'
data = pd.read_csv(file_path)

# Calculate average values for OOSR and Completion Rate
data['Average_OOSR'] = data[['Average_OOSR_Male', 'Average_OOSR_Female']].mean(axis=1)
data['Average_Completion_Rate'] = data[['Average_Completion_Rate_Male', 'Average_Completion_Rate_Female']].mean(axis=1)

# Select relevant columns for the radar chart
radar_data = data[['Country', 'Average_OOSR', 'Average_Completion_Rate',
                   'Lower_Secondary_End_Proficiency_Reading', 'Lower_Secondary_End_Proficiency_Math',
                   'Ladder score', 'Logged GDP per capita', 'Social support',
                   'Healthy life expectancy', 'Freedom to make life choices']]

# Normalize the data for radar chart
scaler = MinMaxScaler()
radar_data_normalized = radar_data.copy()
radar_data_normalized.iloc[:, 1:] = scaler.fit_transform(radar_data.iloc[:, 1:])

# Function to create radar chart
def create_radar_chart(country):
    labels = radar_data.columns[1:]
    num_vars = len(labels)
    
    # Create a 360 degree circle
    angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()
    
    # The radar chart is circular, so we need to "complete the loop" and append the start to the end.
    angles += angles[:1]
    
    # Select the normalized data for the chosen country
    values_normalized = radar_data_normalized[radar_data_normalized['Country'] == country].iloc[0, 1:].tolist()
    values_normalized += values_normalized[:1]
    
    # Select the actual data values for the chosen country for reference
    values_actual = radar_data[radar_data['Country'] == country].iloc[0, 1:].tolist()
    
    fig, ax = plt.subplots(figsize=(8, 8), subplot_kw=dict(polar=True))
    
    ax.fill(angles, values_normalized, color='skyblue', alpha=0.4)
    ax.plot(angles, values_normalized, color='dodgerblue', linewidth=2)
    
    ax.set_yticklabels([])
    ax.set_xticks(angles[:-1])
    ax.set_xticklabels(labels, fontsize=12)
    
    for i, label in enumerate(labels):
        angle_rad = angles[i]
        ha = "right"
        if angle_rad < np.pi / 2 or angle_rad > 3 * np.pi / 2:
            ha = "left"
        ax.text(angle_rad, values_normalized[i], f"{values_actual[i]:.2f}", size=10, horizontalalignment=ha, verticalalignment="center", color="black")
    
    ax.set_title(f'Radar Chart for {country}', size=18, color='dodgerblue', y=1.1)
    
    plt.subplots_adjust(top=0.85, bottom=0.15)
    plt.show()

# Dropdown menu to select country
countries = radar_data['Country'].unique()

country_dropdown = widgets.Dropdown(
    options=countries,
    description='Country:',
    disabled=False,
)

def update_plot(change):
    clear_output(wait=True)
    display(country_dropdown)
    create_radar_chart(change['new'])

country_dropdown.observe(update_plot, names='value')

# Display the initial layout
display(country_dropdown)
create_radar_chart(countries[0])
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from IPython.display import display, clear_output
import ipywidgets as widgets

# Load the dataset
file_path = 'filtered_global_education_world_happiness_average.csv'  # Replace with your local file path
data = pd.read_csv(file_path)

# Calculate average values for OOSR and Completion Rate
data['Average_OOSR'] = data[['Average_OOSR_Male', 'Average_OOSR_Female']].mean(axis=1)
data['Average_Completion_Rate'] = data[['Average_Completion_Rate_Male', 'Average_Completion_Rate_Female']].mean(axis=1)

# Rename columns for clarity
column_mapping = {
    'Average_OOSR': 'Average OOSR',
    'Average_Completion_Rate': 'Average Completion Rate',
    'Lower_Secondary_End_Proficiency_Reading': 'Reading Proficiency',
    'Lower_Secondary_End_Proficiency_Math': 'Math Proficiency'
}

# Apply the renaming
data = data.rename(columns=column_mapping)

# Select relevant columns for the radar chart
radar_data = data[['Country', 'Average OOSR', 'Average Completion Rate',
                   'Reading Proficiency', 'Math Proficiency',
                   'Ladder score', 'Logged GDP per capita', 'Social support',
                   'Healthy life expectancy', 'Freedom to make life choices']]

# Normalize the data for radar chart
scaler = MinMaxScaler()
radar_data_normalized = radar_data.copy()
radar_data_normalized.iloc[:, 1:] = scaler.fit_transform(radar_data.iloc[:, 1:])

# Function to create radar chart for two countries
def create_radar_chart(country1, country2):
    labels = radar_data.columns[1:]
    num_vars = len(labels)
    
    # Create a 360 degree circle
    angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()
    
    # The radar chart is circular, so we need to "complete the loop" and append the start to the end.
    angles += angles[:1]
    
    # Select the normalized data for the chosen countries
    values_normalized_1 = radar_data_normalized[radar_data_normalized['Country'] == country1].iloc[0, 1:].tolist()
    values_normalized_1 += values_normalized_1[:1]
    
    values_normalized_2 = radar_data_normalized[radar_data_normalized['Country'] == country2].iloc[0, 1:].tolist()
    values_normalized_2 += values_normalized_2[:1]
    
    # Select the actual data values for the chosen countries for reference
    values_actual_1 = radar_data[radar_data['Country'] == country1].iloc[0, 1:].tolist()
    values_actual_2 = radar_data[radar_data['Country'] == country2].iloc[0, 1:].tolist()
    
    fig, ax = plt.subplots(figsize=(10, 8), subplot_kw=dict(polar=True))
    
    ax.fill(angles, values_normalized_1, color='skyblue', alpha=0.4, label=country1)
    ax.plot(angles, values_normalized_1, color='dodgerblue', linewidth=2)
    
    ax.fill(angles, values_normalized_2, color='orange', alpha=0.4, label=country2)
    ax.plot(angles, values_normalized_2, color='darkorange', linewidth=2)
    
    ax.set_yticklabels([])
    ax.set_xticks(angles[:-1])
    ax.set_xticklabels(labels, fontsize=12)
    
    # Position the labels outside the plot
    for label, angle in zip(ax.get_xticklabels(), angles):
        x, y = label.get_position()
        if angle == 0:
            label.set_position((x, y + 0.3))
        elif angle == np.pi:
            label.set_position((x, y - 0.3))
        elif angle < np.pi:
            label.set_position((x + 0.2, y + 0.2))
        else:
            label.set_position((x - 0.2, y - 0.2))
    
    ax.legend(loc='upper left', bbox_to_anchor=(1.05, 1))
    
    ax.set_title(f'Radar Chart Comparison: {country1} vs {country2}', size=18, color='dodgerblue', y=1.1)
    
    plt.subplots_adjust(left=0.1, right=0.85, top=0.85, bottom=0.15)
    plt.show()

    # Print the actual values for both countries
    print(f"\nActual Values for {country1}:")
    for label, value in zip(labels, values_actual_1):
        print(f"{label}: {value:.2f}")
    
    print(f"\nActual Values for {country2}:")
    for label, value in zip(labels, values_actual_2):
        print(f"{label}: {value:.2f}")

# Dropdown menus to select countries
countries = radar_data['Country'].unique()

country1_dropdown = widgets.Dropdown(
    options=countries,
    description='Country 1:',
    disabled=False,
)

country2_dropdown = widgets.Dropdown(
    options=countries,
    description='Country 2:',
    disabled=False,
)

def update_plot(change):
    clear_output(wait=True)
    display(widgets.VBox([widgets.HBox([country1_dropdown, country2_dropdown]), widgets.interactive_output(create_radar_chart, {'country1': country1_dropdown, 'country2': country2_dropdown})]))

country1_dropdown.observe(update_plot, names='value')
country2_dropdown.observe(update_plot, names='value')

# Display the initial layout
display(widgets.VBox([widgets.HBox([country1_dropdown, country2_dropdown]), widgets.interactive_output(create_radar_chart, {'country1': country1_dropdown, 'country2': country2_dropdown})]))

# Generate the initial radar chart for the first two countries in the dropdown
create_radar_chart(countries[0], countries[1])
```

```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
from sklearn.preprocessing import MinMaxScaler
import ipywidgets as widgets
from IPython.display import display, clear_output

# Load the dataset
file_path = 'filtered_global_education_world_happiness_average.csv'
data = pd.read_csv(file_path)

# Calculate average values for OOSR and Completion Rate
data['Average_OOSR'] = data[['Average_OOSR_Male', 'Average_OOSR_Female']].mean(axis=1)
data['Average_Completion_Rate'] = data[['Average_Completion_Rate_Male', 'Average_Completion_Rate_Female']].mean(axis=1)

# Rename columns for clarity
column_mapping = {
    'Average_OOSR': 'Average OOSR',
    'Average_Completion_Rate': 'Average Completion Rate',
    'Lower_Secondary_End_Proficiency_Reading': 'Reading Proficiency',
    'Lower_Secondary_End_Proficiency_Math': 'Math Proficiency'
}

# Apply the renaming
data = data.rename(columns=column_mapping)

# Select relevant columns for the radar chart
radar_data = data[['Country', 'Average OOSR', 'Average Completion Rate',
                   'Reading Proficiency', 'Math Proficiency',
                   'Ladder score', 'Logged GDP per capita', 'Social support',
                   'Healthy life expectancy', 'Freedom to make life choices']]

# Normalize the data for radar chart
scaler = MinMaxScaler()
radar_data_normalized = radar_data.copy()
radar_data_normalized.iloc[:, 1:] = scaler.fit_transform(radar_data.iloc[:, 1:])

# Function to create radar chart for two countries
def create_radar_chart(country1, country2):
    labels = radar_data.columns[1:]
    
    # Select the normalized data for the chosen countries
    values_normalized_1 = radar_data_normalized[radar_data_normalized['Country'] == country1].iloc[0, 1:]
    values_normalized_2 = radar_data_normalized[radar_data_normalized['Country'] == country2].iloc[0, 1:]
    
    fig = go.Figure()

    fig.add_trace(go.Scatterpolar(
        r=values_normalized_1,
        theta=labels,
        fill='toself',
        name=country1
    ))
    
    fig.add_trace(go.Scatterpolar(
        r=values_normalized_2,
        theta=labels,
        fill='toself',
        name=country2
    ))

    fig.update_layout(
        polar=dict(
            radialaxis=dict(
                visible=True,
                range=[0, 1]
            )),
        showlegend=True,
        title=f'Radar Chart Comparison: {country1} vs {country2}'
    )

    return fig

# Dropdown menus to select countries
countries = radar_data['Country'].unique()

country1_dropdown = widgets.Dropdown(
    options=countries,
    description='Country 1:',
    disabled=False,
)

country2_dropdown = widgets.Dropdown(
    options=countries,
    description='Country 2:',
    disabled=False,
)

plot_output = widgets.Output()

def update_plot(change):
    with plot_output:
        clear_output(wait=True)
        fig = create_radar_chart(country1_dropdown.value, country2_dropdown.value)
        fig.show()

country1_dropdown.observe(update_plot, names='value')
country2_dropdown.observe(update_plot, names='value')

# Display the initial layout and plot
initial_fig = create_radar_chart(countries[0], countries[1])
plot_output.append_display_data(initial_fig)

display(widgets.VBox([widgets.HBox([country1_dropdown, country2_dropdown]), plot_output]))
```

```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go
from sklearn.preprocessing import MinMaxScaler

# Load the dataset
file_path = 'filtered_global_education_world_happiness_average.csv'
data = pd.read_csv(file_path)

# Calculate average values for OOSR and Completion Rate
data['Average_OOSR'] = data[['Average_OOSR_Male', 'Average_OOSR_Female']].mean(axis=1)
data['Average_Completion_Rate'] = data[['Average_Completion_Rate_Male', 'Average_Completion_Rate_Female']].mean(axis=1)

# Rename columns for clarity
column_mapping = {
    'Average_OOSR': 'Average OOSR',
    'Average_Completion_Rate': 'Average Completion Rate',
    'Lower_Secondary_End_Proficiency_Reading': 'Reading Proficiency',
    'Lower_Secondary_End_Proficiency_Math': 'Math Proficiency'
}

# Apply the renaming
data = data.rename(columns=column_mapping)

# Select relevant columns for the radar chart
radar_data = data[['Country', 'Average OOSR', 'Average Completion Rate',
                   'Reading Proficiency', 'Math Proficiency',
                   'Ladder score', 'Logged GDP per capita', 'Social support',
                   'Healthy life expectancy', 'Freedom to make life choices']]

# Normalize the data for radar chart
scaler = MinMaxScaler()
radar_data_normalized = radar_data.copy()
radar_data_normalized.iloc[:, 1:] = scaler.fit_transform(radar_data.iloc[:, 1:])

# Get the list of countries
countries = radar_data['Country'].unique()

# Function to create radar chart traces
def get_radar_chart_traces(country1, country2):
    labels = radar_data.columns[1:]
    values_normalized_1 = radar_data_normalized[radar_data_normalized['Country'] == country1].iloc[0, 1:]
    values_normalized_2 = radar_data_normalized[radar_data_normalized['Country'] == country2].iloc[0, 1:]

    trace1 = go.Scatterpolar(
        r=values_normalized_1,
        theta=labels,
        fill='toself',
        name=country1
    )
    
    trace2 = go.Scatterpolar(
        r=values_normalized_2,
        theta=labels,
        fill='toself',
        name=country2
    )

    return [trace1, trace2]

# Initialize the figure with the first comparison
initial_traces = get_radar_chart_traces(countries[0], countries[1])
fig = go.Figure(data=initial_traces)

# Create dropdown menu items for the first country
dropdown_buttons_1 = []
for country in countries:
    dropdown_buttons_1.append(
        dict(
            args=[{'r': [radar_data_normalized[radar_data_normalized['Country'] == country].iloc[0, 1:], []],
                   'name': [country, '']}],
            label=country,
            method='restyle'
        )
    )

# Create dropdown menu items for the second country
dropdown_buttons_2 = []
for country in countries:
    dropdown_buttons_2.append(
        dict(
            args=[{'r': [[], radar_data_normalized[radar_data_normalized['Country'] == country].iloc[0, 1:]],
                   'name': ['', country]}],
            label=country,
            method='restyle'
        )
    )

fig.update_layout(
    updatemenus=[
        dict(
            buttons=dropdown_buttons_1,
            direction="down",
            showactive=True,
            x=0.17,
            xanchor='left',
            y=1.15,
            yanchor='top'
        ),
        dict(
            buttons=dropdown_buttons_2,
            direction="down",
            showactive=True,
            x=0.32,
            xanchor='left',
            y=1.15,
            yanchor='top'
        )
    ],
    title='Radar Chart Comparison',
    polar=dict(
        radialaxis=dict(
            visible=True,
            range=[0, 1]
        )
    ),
    showlegend=True
)

fig.show()
```
